```python
# Reviews
playstore.Reviews.unique()
```

```python
# Review
# inside review there is a value with 3.0M with M stand for million, lets change it so it can be measure as float
# replace the specified values
playstore['Reviews'] = playstore.Reviews.replace("0.0",0)
playstore['Reviews'] = playstore.Reviews.replace("3.0M",3000000.0)
playstore['Reviews'] = playstore['Reviews'].astype(float)
sns.distplot(playstore.Reviews)
playstore['Reviews'].dtype

```

```python
playstore.info()
#sns.distplot('Reviews')
```

```python
# Total reviews
plt.figure(figsize=(12,12))
sum_rew = playstore.groupby(['Category'])['Reviews'].sum().sort_values(ascending=False)
sns.barplot(x=sum_rew, y=sum_rew.index, data=playstore)
```

#### Showing the amount of total reviews.

```python
# Mean reviews
plt.figure(figsize=(12,12))
mean_rew = playstore.groupby(['Category'])['Reviews'].mean().sort_values(ascending=False)
sns.barplot(x=mean_rew, y=mean_rew.index, data=playstore)
```

##### This is the average of reviews on each category

```python
# Installs
playstore.Installs.unique()
```

```python
playstore['Installs'] = playstore.Installs.str.replace(",","")
playstore['Installs'] = playstore.Installs.str.replace("+","")
playstore['Installs'] =playstore.Installs.replace("Free",np.nan)
playstore['Installs'] = playstore['Installs'].astype(float)
```

```python
# Total Installs
plt.figure(figsize=(12,12))
sum_inst = playstore.groupby(['Category'])['Installs'].sum().sort_values(ascending=False)
sns.barplot(x=sum_inst, y=sum_inst.index, data=playstore)
```

```python
# Mean Install
plt.figure(figsize=(12,12))
mean_ints = playstore.groupby(['Category'])['Installs'].mean().sort_values(ascending=False)
sns.barplot(x=mean_ints, y=mean_ints.index, data=playstore)
```

```python
# Type for category
playstore.Type.unique()
```

```python
playstore['Type'].replace(to_replace=['0'], value=['Free'],inplace=True)
playstore['Type'].fillna('Free', inplace=True)

print(playstore.groupby('Category')['Type'].value_counts())
Type_cat = playstore.groupby('Category')['Type'].value_counts().unstack().plot.barh(figsize=(10,20), width=0.7)
plt.show()
```

#### So again, family category has the most free and paid app on the google play store. We can see social app is always free, like entertainment, event, education, comic, and more.
#### The medical has a high amount of paid app considering quantity of medical app is not much.

```python
# Android Version
playstore['Android Ver'].unique()
```

#### Now I am going to group it to 1 till 8 version of android. Change the null value to 1.0.

```python
playstore['Android Ver'].replace(to_replace=['4.4W and up','Varies with device'], value=['4.4','1.0'],inplace=True)
playstore['Android Ver'].replace({k: '1.0' for k in ['1.0','1.0 and up','1.5 and up','1.6 and up']},inplace=True)
playstore['Android Ver'].replace({k: '2.0' for k in ['2.0 and up','2.0.1 and up','2.1 and up','2.2 and up','2.2 - 7.1.1','2.3 and up','2.3.3 and up']},inplace=True)
playstore['Android Ver'].replace({k: '3.0' for k in ['3.0 and up','3.1 and up','3.2 and up']},inplace=True)
playstore['Android Ver'].replace({k: '4.0' for k in ['4.0 and up','4.0.3 and up','4.0.3 - 7.1.1','4.1 and up','4.1 - 7.1.1','4.2 and up','4.3 and up','4.4','4.4 and up']},inplace=True)
playstore['Android Ver'].replace({k: '5.0' for k in ['5.0 - 6.0','5.0 - 7.1.1','5.0 - 8.0','5.0 and up','5.1 and up']},inplace=True)
playstore['Android Ver'].replace({k: '6.0' for k in ['6.0 and up']},inplace=True)
playstore['Android Ver'].replace({k: '7.0' for k in ['7.0 - 7.1.1','7.0 and up','7.1 and up']},inplace=True)
playstore['Android Ver'].replace({k: '8.0' for k in ['8.0 and up']},inplace=True)
playstore['Android Ver'].fillna('1.0', inplace=True)
```

```python
print(playstore.groupby('Category')['Android Ver'].value_counts())
Type_cat = playstore.groupby('Category')['Android Ver'].value_counts().unstack().plot.barh(figsize=(10,18), width=1)
plt.show()
```

### Effect of size on rating

#### Make a joinplot to understand the effect of size on rating

```python
sns.jointplot(y ='Size', x ='Rating', data = playstore, kind ='hex')
plt.show()
```

##### Do you see any patterns?
##### Yes, patterns can be observed between Size and Rating ie. their is correlation between Size and Rating.

##### How do you explain the pattern?
##### Generally on increasing Rating, Size of App also increases. But this is not always true ie. for higher Rating, their is constant Size. Thus we can conclude that their is positive correlation between Size and Rating.

### Effect of price on rating

#### Make a jointplot (with regression line)

```python
sns.jointplot(x='Price', y='Rating', data=playstore, kind='reg')
plt.show()
```

#####  What pattern do you see?
##### Generally on increasing the Price, Rating remains almost constant greater than 4.

##### How do you explain the pattern?
##### Since on increasing the Price, Rating remains almost constant greater than 4. Thus it can be concluded that their is very weak Positive correlation between Rating and Price.

##### Replot the data, this time with only records with price > 0

```python
ps1=playstore.loc[playstore.Price>0]
sns.jointplot(x='Price', y='Rating', data=playstore, kind='reg')
plt.show()
```

#### Does the pattern change?
#### Yes, On limiting the record with Price > 0, the overall pattern changed a slight ie their is very weakly Negative Correlation between Price and Rating.
